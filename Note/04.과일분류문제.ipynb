```python
import pandas as pd

fruits = pd.read_csv("../Data/fruits.csv")
fruits.head()
```

```python
fruits.info()
```

```python
# 사과 정보 불러오기
apple = fruits[:35]
apple.tail()
```

#### apple의 산포도

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
	apple.length,
	apple.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title("Apple")

plt.show()
```

```python
# berry 정보 불러오기
berry = fruits[35:]
berry.head()
```

```python
# berry의 산포도
plt.scatter(
	berry.length,
	berry.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title("Berry")

plt.show()
```

```python
# fruits로 산포도
plt.scatter(
	fruits.length,
	fruits.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')
plt.title("fruits")

plt.show()
```

```python
# 2개의 Dataset를 산포도로 한꺼번에 비교하기
plt.scatter(
	apple.length,
	apple.weight
)

plt.scatter(
	berry.length,
	berry.weight
)

plt.xlabel('Length')
plt.ylabel('Weight')

plt.show()
```
