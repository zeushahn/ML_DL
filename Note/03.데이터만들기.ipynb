#### DataFrame 만들기

```python
import pandas as pd
```

```python
apple = \
	pd.DataFrame(
		{
			'length' : apple_length,
			'weight' : apple_weight,
			'name' : 'apple'
		}
	)
apple.head()
```

```python
apple.info()
```

```python
berry = \
	pd.DataFrame(
		{
			'length' : berry_length,
			'weight' : berry_weight,
			'name' : 'berry'
		}
	)
berry.head()
```

```python
berry.info()
```

#### Data 합치기

```python
fruits = \
	pd.concat(
		[apple, berry],
		ignore_index=True # 기존 인덱스를 버리고 새로 시작
	)
fruits.head()
```

```python
fruits.info()
```

#### 저장하기

```python
fruits.to_csv("../Data/fruits.csv", index=False)
```
